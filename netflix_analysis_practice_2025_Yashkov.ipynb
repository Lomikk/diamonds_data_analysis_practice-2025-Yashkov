```python
df = df.drop(df.columns[0], axis=1)
```

```python
# 1.      
df.columns = [col.lower() for col in df.columns]

# 2.    (object)    
text_cols = df.select_dtypes(include="object").columns
for col in text_cols:
    df[col] = df[col].str.lower()
```

```python
df
```

```python
#  
print(":", df.shape)

#      x, y  z
df = df[(df['x'] != 0) & (df['y'] != 0) & (df['z'] != 0)]

#  
print(":", df.shape)
```

```python
df = df.astype({
    'carat'  : 'float32',
    'depth'  : 'float32',
    'table'  : 'float32',
    'x'      : 'float32',
    'y'      : 'float32',
    'z'      : 'float32',
    'price'  : 'int32',
    'cut'    : 'category',
    'color'  : 'category',
    'clarity': 'category'
})
df.info()
```
